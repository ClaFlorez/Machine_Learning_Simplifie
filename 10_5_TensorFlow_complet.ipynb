```python
#Premier projet TensorFlow complet
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Votre premier projet Deep Learning avec TensorFlow
print("PREMIER PROJET DEEP LEARNING AVEC TENSORFLOW")
print("Objectif: Classification d'images de vêtements (Fashion-MNIST)")
print("=" * 80)

# Vérifier l'installation de TensorFlow
print(f"Version TensorFlow: {tf.__version__}")
print(f"GPU disponible: {tf.config.list_physical_devices('GPU')}")

# Charger le dataset Fashion-MNIST (inclus dans TensorFlow)
print(f"\nChargement du dataset Fashion-MNIST:")
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Classes de vêtements
class_names = ['T-shirt/top', 'Pantalon', 'Pull', 'Robe', 'Manteau',
               'Sandale', 'Chemise', 'Sneaker', 'Sac', 'Bottine']

print(f"  • Images d'entraînement: {X_train.shape[0]:,}")
print(f"  • Images de test: {X_test.shape[0]:,}")
print(f"  • Taille des images: {X_train.shape[1]}x{X_train.shape[2]} pixels")
print(f"  • Classes: {len(class_names)} types de vêtements")
print(f"  • Classes: {class_names}")

# Exploration des données
print(f"\nExploration des données:")
print("=" * 40)

# Distribution des classes
unique, counts = np.unique(y_train, return_counts=True)
print(f"Distribution des classes d'entraînement:")
for class_id, count in zip(unique, counts):
    print(f"  • {class_names[class_id]}: {count:,} images ({count/len(y_train)*100:.1f}%)")

# Visualiser quelques exemples
plt.figure(figsize=(15, 10))

for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f'{class_names[y_train[i]]}')
    plt.axis('off')

plt.suptitle('Exemples du Dataset Fashion-MNIST', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Préprocessing des données
print(f"\nPréparation des données pour Deep Learning:")
print("=" * 50)

# Normaliser les pixels (0-255 → 0-1)
X_train_normalized = X_train.astype('float32') / 255.0
X_test_normalized = X_test.astype('float32') / 255.0

print(f"Normalisation des pixels:")
print(f"  • Avant: pixels entre {X_train.min()} et {X_train.max()}")
print(f"  • Après: pixels entre {X_train_normalized.min():.1f} et {X_train_normalized.max():.1f}")

# Reshape pour le réseau (ajouter dimension channel)
X_train_reshaped = X_train_normalized.reshape(-1, 28, 28, 1)
X_test_reshaped = X_test_normalized.reshape(-1, 28, 28, 1)

print(f"Reshape des données:")
print(f"  • Forme originale: {X_train.shape}")
print(f"  • Forme pour CNN: {X_train_reshaped.shape}")

# Construire le modèle CNN avec Keras
print(f"\nConstruction du modèle CNN:")
print("=" * 40)

model = keras.Sequential([
    # Première couche de convolution
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    keras.layers.MaxPooling2D((2, 2)),

    # Deuxième couche de convolution
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),

    # Troisième couche de convolution
    keras.layers.Conv2D(64, (3, 3), activation='relu'),

    # Aplatir pour les couches denses
    keras.layers.Flatten(),

    # Couches denses finales
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.5),  # Régularisation
    keras.layers.Dense(10, activation='softmax')  # 10 classes
])

# Afficher l'architecture
print("Architecture du modèle CNN:")
model.summary()

# Compiler le modèle
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(f"\nConfiguration de l'entraînement:")
print(f"  • Optimiseur: Adam (adaptatif)")
print(f"  • Fonction de perte: Categorical Crossentropy")
print(f"  • Métrique: Accuracy")

# Entraîner le modèle
print(f"\nEntraînement du modèle CNN:")
print("=" * 40)

# Créer un set de validation
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train_reshaped, y_train, test_size=0.2, random_state=42
)

print(f"Division finale des données:")
print(f"  • Entraînement: {len(X_train_final):,} images")
print(f"  • Validation: {len(X_val):,} images")
print(f"  • Test: {len(X_test):,} images")

# Entraîner avec callbacks pour monitoring
callbacks = [
    keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.5)
]

history = model.fit(
    X_train_final, y_train_final,
    batch_size=32,
    epochs=20,
    validation_data=(X_val, y_val),
    callbacks=callbacks,
    verbose=1
)

# Évaluer sur le test final
test_loss, test_accuracy = model.evaluate(X_test_reshaped, y_test, verbose=0)

print(f"\nRésultats finaux:")
print(f"  • Accuracy sur test: {test_accuracy:.4f} ({test_accuracy:.1%})")
print(f"  • Loss sur test: {test_loss:.4f}")

# Analyser l'entraînement
plt.figure(figsize=(15, 10))

# Courbes d'entraînement
plt.subplot(2, 2, 1)
plt.plot(history.history['accuracy'], label='Accuracy Entraînement', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Accuracy Validation', linewidth=2)
plt.title('Évolution de l\'Accuracy', fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(history.history['loss'], label='Loss Entraînement', linewidth=2)
plt.plot(history.history['val_loss'], label='Loss Validation', linewidth=2)
plt.title('Évolution de la Loss', fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Prédictions sur quelques exemples
predictions = model.predict(X_test_reshaped[:12])
predicted_classes = np.argmax(predictions, axis=1)

# Visualiser les prédictions
for i in range(12):
    plt.subplot(4, 6, i + 13)  # Commencer après les 2 graphiques précédents
    plt.imshow(X_test[i], cmap='gray')

    predicted_class = predicted_classes[i]
    true_class = y_test[i]
    confidence = predictions[i][predicted_class]

    if predicted_class == true_class:
        color = 'green'
        status = 'CORRECT'
    else:
        color = 'red'
        status = 'ERREUR'

    plt.title(f'{status}\nPrédit: {class_names[predicted_class]}\nVrai: {class_names[true_class]}\nConf: {confidence:.2f}',
             color=color, fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()

# Analyser les erreurs
print(f"\nAnalyse des erreurs de classification:")
print("=" * 50)

# Matrice de confusion
from sklearn.metrics import confusion_matrix, classification_report

all_predictions = model.predict(X_test_reshaped)
all_predicted_classes = np.argmax(all_predictions, axis=1)

# Rapport de classification
print("Rapport de classification détaillé:")
report = classification_report(y_test, all_predicted_classes,
                             target_names=class_names, output_dict=True)

for class_name, metrics in report.items():
    if class_name not in ['accuracy', 'macro avg', 'weighted avg']:
        if isinstance(metrics, dict):
            print(f"\n{class_name}:")
            print(f"  • Précision: {metrics['precision']:.3f}")
            print(f"  • Rappel: {metrics['recall']:.3f}")
            print(f"  • F1-score: {metrics['f1-score']:.3f}")
            print(f"  • Support: {metrics['support']}")

# Identifier les confusions les plus fréquentes
cm = confusion_matrix(y_test, all_predicted_classes)
print(f"\nConfusions les plus fréquentes:")

confusions = []
for i in range(len(class_names)):
    for j in range(len(class_names)):
        if i != j and cm[i, j] > 20:  # Plus de 20 confusions
            confusions.append((class_names[i], class_names[j], cm[i, j]))

confusions.sort(key=lambda x: x[2], reverse=True)

for vraie_classe, classe_predite, nb_erreurs in confusions[:5]:
    print(f"  • {vraie_classe} confondu avec {classe_predite}: {nb_erreurs} fois")

# Sauvegarder le modèle
print(f"\nSauvegarde du modèle:")
print("=" * 30)

model.save('modele_fashion_mnist.h5')
print("✓ Modèle sauvegardé: modele_fashion_mnist.h5")

# Charger et tester le modèle sauvegardé
modele_charge = keras.models.load_model('modele_fashion_mnist.h5')
test_accuracy_loaded = modele_charge.evaluate(X_test_reshaped, y_test, verbose=0)[1]

print(f"✓ Modèle rechargé avec succès")
print(f"✓ Accuracy confirmée: {test_accuracy_loaded:.4f}")

# Déploiement et utilisation
print(f"\nUtilisation du modèle en production:")
print("=" * 50)

def classifier_vetement(image_path_ou_array):
    """
    Fonction pour classifier un vêtement en production
    """
    # Si c'est un array numpy (notre cas de test)
    if isinstance(image_path_ou_array, np.ndarray):
        image = image_path_ou_array
    else:
        # Charger l'image depuis un fichier (cas réel)
        image = keras.preprocessing.image.load_img(image_path_ou_array,
                                                  target_size=(28, 28),
                                                  color_mode='grayscale')
        image = keras.preprocessing.image.img_to_array(image)

    # Préprocessing identique à l'entraînement
    image = image.astype('float32') / 255.0
    image = image.reshape(1, 28, 28, 1)

    # Prédiction
    prediction = modele_charge.predict(image, verbose=0)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class]

    return {
        'classe': class_names[predicted_class],
        'confiance': confidence,
        'probabilites': {class_names[i]: prediction[0][i] for i in range(len(class_names))}
    }

# Tester la fonction de classification
print(f"Test de la fonction de classification:")
print("-" * 50)

# Prendre quelques images de test
for i in [0, 100, 500, 1000]:
    resultat = classifier_vetement(X_test[i])
    vraie_classe = class_names[y_test[i]]

    print(f"\nImage {i}:")
    print(f"  • Vraie classe: {vraie_classe}")
    print(f"  • Prédiction: {resultat['classe']}")
    print(f"  • Confiance: {resultat['confiance']:.1%}")

    if resultat['classe'] == vraie_classe:
        print(f"  ✓ Classification correcte")
    else:
        print(f"  ✗ Erreur de classification")

        # Afficher les top 3 prédictions
        probs_sorted = sorted(resultat['probabilites'].items(),
                            key=lambda x: x[1], reverse=True)
        print(f"  Top 3 prédictions:")
        for classe, prob in probs_sorted[:3]:
            print(f"    - {classe}: {prob:.1%}")

# Optimisation et bonnes pratiques
print(f"\nOptimisation et bonnes pratiques TensorFlow:")
print("=" * 60)

print("PERFORMANCE:")
print("  • Utilisez tf.data pour charger les données efficacement")
print("  • Activez le mixed precision training (GPU)")
print("  • Optimisez la taille des batches")
print("  • Utilisez la compilation XLA si possible")

print("\nMONITORING:")
print("  • TensorBoard pour visualiser l'entraînement")
print("  • Callbacks pour early stopping")
print("  • Sauvegarde régulière des checkpoints")
print("  • Logging des métriques importantes")

print("\nDÉPLOIEMENT:")
print("  • TensorFlow Serving pour APIs REST")
print("  • TensorFlow Lite pour mobile")
print("  • TensorFlow.js pour navigateur web")
print("  • Docker pour containerisation")

# Projet d'amélioration
print(f"\nPistes d'amélioration du modèle:")
print("=" * 50)

print("ARCHITECTURE:")
print("  • Ajouter plus de couches convolutives")
print("  • Utiliser des architectures pré-entraînées")
print("  • Implémenter des connexions résiduelles")
print("  • Expérimenter avec différentes fonctions d'activation")

print("\nDONNÉES:")
print("  • Augmentation de données (rotation, zoom, etc.)")
print("  • Collecte de données supplémentaires")
print("  • Nettoyage des données bruitées")
print("  • Équilibrage des classes")

print("\nENTRAÎNEMENT:")
print("  • Ajustement du learning rate")
print("  • Utilisation de différents optimiseurs")
print("  • Régularisation (dropout, L2)")
print("  • Entraînement plus long avec patience")

# Métriques de production
print(f"\nMétriques pour le monitoring en production:")
print("=" * 50)

print("PERFORMANCE MODÈLE:")
print("  • Accuracy globale > 90%")
print("  • Précision par classe > 85%")
print("  • Temps de prédiction < 100ms")
print("  • Utilisation mémoire < 500MB")

print("\nQUALITÉ DONNÉES:")
print("  • Détection de drift dans les images")
print("  • Monitoring de la distribution des classes")
print("  • Détection d'images aberrantes")
print("  • Validation de la qualité d'image")

print("\nBUSINESS:")
print("  • Taux de satisfaction utilisateur")
print("  • Temps de traitement par image")
print("  • Coût par prédiction")
print("  • ROI du système automatisé")

print(f"\n🚀 Félicitations!")
print("Vous venez de créer votre premier modèle Deep Learning")
print("avec TensorFlow ! C'est le début d'une aventure passionnante")
print("dans le monde de l'intelligence artificielle moderne.")
```
