```python
# Analyser des avis clients (donn√©es non-structur√©es)
import pandas as pd
from collections import Counter
import re

# Exemples d'avis clients (donn√©es non-structur√©es)
avis_clients = [
    "Ce produit est fantastique ! Je le recommande vivement. Excellent service client.",
    "Tr√®s d√©√ßu de mon achat. La qualit√© n'est pas au rendez-vous. Service client lent.",
    "Produit correct, rien d'exceptionnel. Livraison rapide. Prix un peu √©lev√©.",
    "Absolument parfait ! D√©passe mes attentes. Bravo √† l'√©quipe !",
    "Mauvaise exp√©rience. Produit d√©fectueux. Remboursement difficile."
]

print("üìù Avis clients originaux (non-structur√©s) :")
for i, avis in enumerate(avis_clients, 1):
    print(f"{i}. {avis}")

# Extraction d'informations structur√©es depuis le texte
def analyser_sentiment(texte):
    mots_positifs = ['fantastique', 'excellent', 'parfait', 'bravo', 'recommande', 'rapide']
    mots_negatifs = ['d√©√ßu', 'lent', 'mauvaise', 'd√©fectueux', 'difficile']

    texte_lower = texte.lower()
    score_positif = sum(1 for mot in mots_positifs if mot in texte_lower)
    score_negatif = sum(1 for mot in mots_negatifs if mot in texte_lower)

    if score_positif > score_negatif:
        return "Positif"
    elif score_negatif > score_positif:
        return "N√©gatif"
    else:
        return "Neutre"

def extraire_longueur(texte):
    return len(texte.split())

# Transformation en donn√©es structur√©es
donnees_structurees = []
for i, avis in enumerate(avis_clients):
    donnees_structurees.append({
        'id_avis': i+1,
        'sentiment': analyser_sentiment(avis),
        'longueur_mots': extraire_longueur(avis),
        'contient_service': 'service' in avis.lower(),
        'contient_produit': 'produit' in avis.lower(),
        'texte_original': avis
    })

# Cr√©er un DataFrame structur√©
df_avis = pd.DataFrame(donnees_structurees)

print("\nüìä Donn√©es transform√©es (structur√©es) :")
print(df_avis[['id_avis', 'sentiment', 'longueur_mots', 'contient_service', 'contient_produit']])

print("\nüìà Analyse des sentiments :")
print(df_avis['sentiment'].value_counts())

print("\nüîç Statistiques :")
print(f"Longueur moyenne des avis : {df_avis['longueur_mots'].mean():.1f} mots")
print(f"Avis mentionnant le service : {df_avis['contient_service'].sum()}")
print(f"Avis mentionnant le produit : {df_avis['contient_produit'].sum()}")
```

```python
# Diagnostic automatique de la qualit√© des donn√©es
import pandas as pd
import numpy as np

# Cr√©er un dataset avec des probl√®mes typiques
data_problematique = {
    'nom': ['Jean Dupont', 'marie martin', 'PIERRE DURAND', None, 'Jean Dupont'],  # Doublons, casse, manquant
    'age': [25, 999, 30, 28, 25],  # Valeur aberrante
    'email': ['jean@gmail.com', 'marie@yahoo', 'pierre@outlook.com', 'invalide', 'jean@gmail.com'],  # Format invalide
    'salaire': [45000, 52000, None, 48000, 45000],  # Valeur manquante
    'ville': ['Paris', 'LYON', 'marseille', 'Paris', 'Paris'],  # Casse incoh√©rente
    'date_embauche': ['2024-01-15', '2023-13-45', '2022-06-10', '2024-02-30', '2024-01-15']  # Dates invalides
}

df = pd.DataFrame(data_problematique)

print("üîç DIAGNOSTIC DE QUALIT√â DES DONN√âES")
print("=" * 50)

# 1. Vue d'ensemble
print("\nüìä Aper√ßu des donn√©es :")
print(df)

print(f"\nüìè Dimensions : {df.shape[0]} lignes, {df.shape[1]} colonnes")

# 2. Valeurs manquantes
print("\n‚ùì VALEURS MANQUANTES :")
valeurs_manquantes = df.isnull().sum()
for colonne, nb_manquant in valeurs_manquantes.items():
    if nb_manquant > 0:
        pourcentage = (nb_manquant / len(df)) * 100
        print(f"   {colonne}: {nb_manquant} ({pourcentage:.1f}%)")

# 3. Doublons
print(f"\nüîÑ DOUBLONS : {df.duplicated().sum()} lignes dupliqu√©es")
if df.duplicated().sum() > 0:
    print("   Lignes dupliqu√©es :")
    print(df[df.duplicated()])

# 4. Valeurs aberrantes num√©riques
print("\nüö® VALEURS ABERRANTES :")
colonnes_numeriques = df.select_dtypes(include=[np.number]).columns
for col in colonnes_numeriques:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_basse = Q1 - 1.5 * IQR
    limite_haute = Q3 + 1.5 * IQR

    aberrantes = df[(df[col] < limite_basse) | (df[col] > limite_haute)][col]
    if len(aberrantes) > 0:
        print(f"   {col}: {list(aberrantes.values)} (limites: {limite_basse:.0f} - {limite_haute:.0f})")

# 5. Probl√®mes de format
print("\nüìù PROBL√àMES DE FORMAT :")

# Emails invalides
emails_invalides = df[~df['email'].str.contains('@.*\\.', na=False)]['email']
if len(emails_invalides) > 0:
    print(f"   Emails invalides : {list(emails_invalides.values)}")

# Incoh√©rences de casse
print("\nüî§ INCOH√âRENCES DE CASSE :")
for col in ['nom', 'ville']:
    if col in df.columns:
        valeurs_uniques = df[col].dropna().unique()
        valeurs_normalisees = set(v.lower().strip() for v in valeurs_uniques if isinstance(v, str))
        if len(valeurs_uniques) > len(valeurs_normalisees):
            print(f"   {col}: {list(valeurs_uniques)} ‚Üí probl√®me de casse d√©tect√©")

print("\nüí° RECOMMANDATIONS :")
print("   1. Supprimer ou imputer les valeurs manquantes")
print("   2. Corriger les valeurs aberrantes (√¢ge = 999)")
print("   3. Standardiser la casse des textes")
print("   4. Valider le format des emails")
print("   5. Supprimer les doublons")
print("   6. V√©rifier les dates invalides")
```

```python
# Validation avanc√©e avec r√®gles m√©tier
def valider_donnees(df):
    """Valide un DataFrame selon des r√®gles m√©tier"""
    erreurs = []

    # R√®gle 1: √Çge entre 16 et 80 ans
    ages_invalides = df[(df['age'] < 16) | (df['age'] > 80)]
    if len(ages_invalides) > 0:
        erreurs.append(f"√Çges invalides: {len(ages_invalides)}")

    # R√®gle 2: Salaire entre 15k et 200k
    salaires_invalides = df[(df['salaire'] < 15000) | (df['salaire'] > 200000)]
    if len(salaires_invalides) > 0:
        erreurs.append(f"Salaires invalides: {len(salaires_invalides)}")

    # R√®gle 3: Email contient @ et .
    emails_invalides = df[~df['email'].str.contains('@.*\\.', na=False)]
    if len(emails_invalides) > 0:
        erreurs.append(f"Emails invalides: {len(emails_invalides)}")

    return erreurs

# Tester la validation
erreurs = valider_donnees(df_propre)
if erreurs:
    print("‚ùå Erreurs d√©tect√©es :")
    for erreur in erreurs:
        print(f"   - {erreur}")
else:
    print("‚úÖ Toutes les validations pass√©es !")
```

Cas d'√©tude : Sauver un projet e-commerce
Une startup e-commerce veut pr√©dire quels clients vont acheter. Leurs donn√©es initiales sont catastrophiques :

üìä DONN√âES INITIALES (PROBL√âMATIQUES) :

client_id | age | email              | achat_precedent | ville    | satisfaction
----------|-----|--------------------|-----------------|-----------|--------------
001       | 25  | jean@gmail.com     | 150.0          | Paris    | 4.5
002       | 999 | marie@             | NaN            | LYON     | 3.0
003       | 30  | pierre@outlook.com | 75.5           | marseille| NaN
004       | NaN | invalide           | 200.0          | Paris    | 5.0
001       | 25  | jean@gmail.com     | 150.0          | Paris    | 4.5

PROBL√àMES D√âTECT√âS :
‚ùå √Çge aberrant (999 ans)
‚ùå Emails invalides
‚ùå Valeurs manquantes (√¢ge, achat, satisfaction)
‚ùå Incoh√©rences de casse (villes)
‚ùå Doublons (client 001)

```python
# Solution compl√®te de nettoyage
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

def nettoyer_donnees_ecommerce(df):
    """Nettoyage complet pour donn√©es e-commerce"""

    print("üßπ √âTAPE 1: Suppression des doublons")
    avant = len(df)
    df = df.drop_duplicates().copy() # Use .copy() to avoid SettingWithCopyWarning
    print(f"   {avant - len(df)} doublons supprim√©s")

    print("\nüî§ √âTAPE 2: Normalisation des textes")
    df.loc[:, 'ville'] = df['ville'].str.title().str.strip()

    print("\nüìß √âTAPE 3: Nettoyage des emails")
    # Marquer les emails invalides
    df.loc[:, 'email_valide'] = df['email'].str.contains('@.*\\.', na=False)

    print("\nüö® √âTAPE 4: Correction des valeurs aberrantes")
    # √Çges impossibles ‚Üí mediane
    median_age = df['age'].median()
    df.loc[df['age'] > 120, 'age'] = median_age
    df.loc[df['age'] < 16, 'age'] = median_age


    print("\nüìä √âTAPE 5: Imputation des valeurs manquantes")
    # Imputation intelligente avec KNN
    colonnes_numeriques = ['age', 'achat_precedent', 'satisfaction']
    imputer = KNNImputer(n_neighbors=2)
    df.loc[:, colonnes_numeriques] = imputer.fit_transform(df[colonnes_numeriques])

    print("\n‚úÖ √âTAPE 6: Validation finale")
    # V√©rifications finales
    assert df.isnull().sum().sum() == 0, "Il reste des valeurs manquantes"
    assert (df['age'] >= 16).all() and (df['age'] <= 120).all(), "√Çges invalides"
    assert (df['satisfaction'] >= 1).all() and (df['satisfaction'] <= 5).all(), "Satisfaction invalide"

    return df

# Create the dataframe from the data in the markdown cell
data_ecommerce = {
    'client_id': ['001', '002', '003', '004', '001'],
    'age': [25, 999, 30, np.nan, 25],
    'email': ['jean@gmail.com', 'marie@', 'pierre@outlook.com', 'invalide', 'jean@gmail.com'],
    'achat_precedent': [150.0, np.nan, 75.5, 200.0, 150.0],
    'ville': ['Paris', 'LYON', 'marseille', 'Paris', 'Paris'],
    'satisfaction': [4.5, 3.0, np.nan, 5.0, 4.5]
}
df_ecommerce = pd.DataFrame(data_ecommerce)


# Appliquer le nettoyage
df_ecommerce_propre = nettoyer_donnees_ecommerce(df_ecommerce)

print("\nüéâ DONN√âES NETTOY√âES :")
print(df_ecommerce_propre)

print(f"\nüìà AM√âLIORATION DE LA QUALIT√â :")
print(f"   Compl√©tude : 100% (√©tait ~60%)")
print(f"   Coh√©rence : 100% (√©tait ~40%)")
print(f"   Validit√© : 100% (√©tait ~30%)")
print(f"   Pr√™t pour le Machine Learning ! ‚úÖ")
```
